```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/processed/covid19_with_predictions.csv')
df['Date'] = pd.to_datetime(df['Date'])

sns.set_theme(style="darkgrid")

print("Data loaded successfully!")
print(df.shape)
df.head()
```

```python
# Plot 1: Total Confirmed Cases Over Time
plt.figure(figsize=(12, 5))
plt.plot(df['Date'], df['Confirmed'], color='steelblue', linewidth=2)
plt.title('Global COVID-19 Confirmed Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Total Confirmed Cases')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plot 2: Daily New Cases Over Time
plt.figure(figsize=(12, 5))
plt.fill_between(df['Date'], df['new_cases'], color='orange', alpha=0.6)
plt.title('Daily New COVID-19 Cases Over Time')
plt.xlabel('Date')
plt.ylabel('New Cases Per Day')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plot 3: Deaths vs Confirmed Cases
plt.figure(figsize=(8, 5))
plt.scatter(df['Confirmed'], df['Deaths'], color='red', alpha=0.3, s=10)
plt.title('Deaths vs Confirmed Cases')
plt.xlabel('Total Confirmed Cases')
plt.ylabel('Total Deaths')
plt.tight_layout()
plt.show()
```

```python
# Plot 4: Actual vs Predicted Confirmed Cases
plt.figure(figsize=(12, 5))
plt.plot(df['Date'], df['Confirmed'], color='steelblue', linewidth=2, label='Actual')
plt.plot(df['Date'], df['predicted_confirmed'], color='red', linewidth=2, linestyle='--', label='Predicted')
plt.title('Actual vs Predicted Confirmed Cases (Linear Regression)')
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
